# Exploração inicial dos dados

```python
def explorar_dataframe(df, nome):
    print(f"Visão geral de {nome}:\n")
    print(df.head(), "\n")
    print(f"Resumo Estatístico de {nome}:\n")
    print(df.describe(include='all'), "\n")
    print(f"Informações de {nome}:\n")
    print(df.info(), "\n")

    # Verificar se o DataFrame possui colunas numéricas antes de plotar
    if not df.select_dtypes(include=[np.number]).empty:
        df.hist(bins=30, figsize=(15, 10))
        plt.suptitle(f'Histograma para {nome}', fontsize=16)
        plt.show()
    else:
        print(f"Nenhuma coluna numérica para plotar em {nome}.\n")
```

#Limpeza de dados

Verificar e tratar valores nulos

```python
# Função para verificar valores nulos
def verificar_nulos(df, nome):
    print(f"Verificando valores nulos no DataFrame '{nome}':")
    print(df.isnull().sum(), "\n")

# Chamar a função para cada DataFrame
verificar_nulos(df_customers, "Customers")
verificar_nulos(df_geolocation, "Geolocation")
verificar_nulos(df_order_items, "Order Items")
verificar_nulos(df_order_payments, "Order Payments")
verificar_nulos(df_order_reviews, "Order Reviews")
verificar_nulos(df_orders, "Orders")
verificar_nulos(df_products, "Products")
verificar_nulos(df_sellers, "Sellers")
verificar_nulos(df_product_translation, "Product Category Translation")
```

Tratando valores nulos

```python
# Função para preencher valores nulos e strings vazias
def clean_missing_data(df, column, fill_value):
    # Substituir espaços em branco e strings vazias por NaN
    df[column] = df[column].replace(r'^\s*$', pd.NA, regex=True)
    # Preencher NaN com o valor especificado
    df[column] = df[column].fillna(fill_value)

# Aplicar a limpeza para campos textuais relevantes
clean_missing_data(df_order_reviews, 'review_comment_title', 'No Title')
clean_missing_data(df_order_reviews, 'review_comment_message', 'No Comment')
clean_missing_data(df_products, 'product_category_name', 'Unknown')
```

Corrigindo tipos de dados

```python
# Corrigir tipos de dados de colunas de datas
df_orders['order_purchase_timestamp'] = pd.to_datetime(df_orders['order_purchase_timestamp'])
df_orders['order_approved_at'] = pd.to_datetime(df_orders['order_approved_at'])
df_orders['order_delivered_carrier_date'] = pd.to_datetime(df_orders['order_delivered_carrier_date'])
df_orders['order_delivered_customer_date'] = pd.to_datetime(df_orders['order_delivered_customer_date'])
df_orders['order_estimated_delivery_date'] = pd.to_datetime(df_orders['order_estimated_delivery_date'])
df_order_reviews['review_creation_date'] = pd.to_datetime(df_order_reviews['review_creation_date'])
df_order_reviews['review_answer_timestamp'] = pd.to_datetime(df_order_reviews['review_answer_timestamp'])
```
